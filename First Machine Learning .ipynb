```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split



iris = load_iris()

X=iris.data
y= iris.target

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=5)
```

```python
#SUPPORT VECTOR MACHINE SVC CLASSIFICATION
from sklearn.svm import SVC
from sklearn import metrics
#from sklearn.metrics import classification_report
#from sklearn.metrics import confusion_matrix


#DEFAULT VALUES
print("___________DEFAULT VALUES FOR SVM___________")
clf = SVC()
clf.fit(X_train,y_train)

#PREDICTION AND ACCURACY TEST
predt = clf.predict(X_test)
print("######Accuracy Test######")
print(metrics.accuracy_score(y_test,predt))
print("\n")

#MATRICES VALUES
print("######Classification Report######")
print(classification_report(y_test, predt))


#CONFUSION MATRIX
print("######Confusion Matrix######")
print(confusion_matrix(y_test, predt))
print("\n\n")


#SVM KERNEL=linear VALUES

print("___________LINEAR KERNEL VALUE FOR SVM___________")
clf = SVC(kernel='linear')
clf.fit(X_train,y_train)

#PREDICTION AND ACCURACY TEST
predt = clf.predict(X_test)
print("######Accuracy Test######")
print(metrics.accuracy_score(y_test,predt))
print("\n")

#MATRICES VALUES
print("######Classification Report######")
print(classification_report(y_test, predt))

#CONFUSION MATRIX
print("######Confusion Matrix######")
print(confusion_matrix(y_test, predt))
print("\n\n")

#SVM KERNEL=polynomial VALUES

print("___________POLYNOMIAL KERNEL VALUE FOR SVM___________")
clf = SVC(kernel='poly')
clf.fit(X_train,y_train)

#PREDICTION AND ACCURACY TEST
predt = clf.predict(X_test)
print("######Accuracy Test######")
print(metrics.accuracy_score(y_test,predt))
print("\n")

#MATRICES VALUES
print("######Classification Report######")
print(classification_report(y_test, predt))

#CONFUSION MATRIX
print("######Confusion Matrix######")
print(confusion_matrix(y_test, predt))
print("\n\n")
```

```python
#DESCION TREE CLASSIFICATION
from sklearn.tree import DecisionTreeClassifier

#DEFAULT VALUES
print("___________DEFAULT VALUES FOR DESCION TREE CLASSIFIER___________")

dt= DecisionTreeClassifier(random_state=0)

dt.fit(X_train,y_train)
predt_two = dt.predict(X_test)

print("######Accuracy Test######")
print(metrics.accuracy_score(y_test,predt_two))
print("\n")


#MATRICES VALUES
print("######Classification Report######")
print(classification_report(y_test, predt))

#CONFUSION MATRIX
print("######Confusion Matrix######")
print(confusion_matrix(y_test, predt))

print("\n\n")


#ENTROPY VALUES
print("___________CRITERION ENTROPI VALUES FOR DESCION TREE CLASSIFIER___________")

dt_entopi= DecisionTreeClassifier(random_state=0,criterion='entropy')

dt_entopi.fit(X_train,y_train)
predt_entopi = dt_entopi.predict(X_test)

print("######Accuracy Test######")
print(metrics.accuracy_score(y_test,predt_entopi))
print("\n")


#MATRICES VALUES
print("######Classification Report######")
print(classification_report(y_test, predt_entopi))

#CONFUSION MATRIX
print("######Confusion Matrix######")
print(confusion_matrix(y_test, predt_entopi))

print("\n\n")

#RANDOM SPLITTER AND MAX_DEPTH=5 VALUES
print("___________SPLITTER:RANDOM AND MAX_DEPTH:5 VALUES FOR DESCION TREE CLASSIFIER___________")

dt_random= DecisionTreeClassifier(random_state=0,criterion='entropy',splitter='random', max_depth=5)

dt_random.fit(X_train,y_train)
predt_random = dt_random.predict(X_test)

print("######Accuracy Test######")
print(metrics.accuracy_score(y_test,predt_random))
print("\n")


#MATRICES VALUES
print("######Classification Report######")
print(classification_report(y_test,predt_random))

#CONFUSION MATRIX
print("######Confusion Matrix######")
print(confusion_matrix(y_test,predt_random))

print("\n\n")
```

```python
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
#DEFAULT VALUES FOR VOTING CLASSIFIER

print("___________DEFAULT VALUES FOR VOTING CLASSIFIER___________")

estimator = []
estimator.append(('LR', 
                  LogisticRegression(solver ='lbfgs', 
                                     multi_class ='multinomial', 
                                     max_iter = 200)))
estimator.append(('SVC', SVC(gamma ='auto', probability = True)))
estimator.append(('DTC', DecisionTreeClassifier()))

vc=VotingClassifier(estimators=estimator, voting ='hard')


vc.fit(X_train, y_train)
result = vc.predict(X_test)

print("######Accuracy Test######")
print(metrics.accuracy_score(y_test, result))
print("\n")


print("######Classification Report######")
print(classification_report(y_test, result))

print("######Confusion Matrix######")
print(confusion_matrix(y_test,result))
print("\n\n")
```
